## Three dimensional plots

```
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
def f1(x, y):
    return (x*2 + y)
def f2(x, y):
    return (y**2 - x)
def f3(x, y):
    return (x**2 + y)
ax = plt.axes(projection='3d')
xline = np.linspace(0, 15, 10)
yline = np.linspace(0, 15, 10)
z1line = f1(xline, yline)
ax.plot3D(xline, yline, z1line, 'gray')
p1x = np.array([0])
p1y = np.array([0])
p1z = np.array([0])
plt.plot(p1x, p1y, p1z, 'ro')
#ax.view_init(35, 15)

z2line = f2(xline, yline)
ax.plot3D(xline, yline, z2line, 'green')
p2x = np.array([15])
p2y = np.array([15])
p2z = np.array([210])
plt.plot(p2x, p2y, p2z, 'ro')

z3line = f3(xline, yline)
ax.plot3D(xline, yline, z3line, 'blue')
p3x = np.array([15])
p3y = np.array([15])
p3z = np.array([240])
plt.plot(p3x, p3y, p3z, 'ro')
```

## Contour plots


```
plt.style.use('seaborn-white')
X, Y = np.meshgrid(xline, yline)
Z1 = f1(X, Y)
plt.contour(X, Y, Z1, colors='black');
```

```
Z2 = f2(X, Y)
plt.contour(X, Y, Z2, colors='black');
```

```
Z3 = f3(X, Y)
plt.contour(X, Y, Z3, colors='black');
```

## Density plots



```
Z1 = f1(X, Y)
plt.contour(X, Y, Z1, 20, cmap='RdGy')
plt.colorbar()
```

```
Z2 = f2(X, Y)
plt.contour(X, Y, Z2, 20, cmap='RdGy')
plt.colorbar()
```

```
Z3 = f3(X, Y)
plt.contour(X, Y, Z3, 20, cmap='RdGy')
plt.colorbar()
```
